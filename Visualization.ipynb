```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.scatterplot(data=train_df,x='floor_area',hue='building_class',y='site_eui')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.scatterplot(data=train_df,y='floor_area',hue='State_Factor',x='energy_star_rating')

plt.subplot(1,2,2)
sns.scatterplot(data=train_df,y='floor_area',hue='building_class',x='energy_star_rating')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(data=train_df, x="ELEVATION", y="site_eui",hue='building_class')
plt.show()
```
