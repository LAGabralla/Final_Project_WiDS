# Cleaning Data (Handling missing values)

```python
# check number of missing values in data

round(train_df.isna().sum()* 100 / len(train_df),2)
```

```python
# #(1) using KNNImputer
# imputer = KNNImputer()
# for col in ['year_built','energy_star_rating','direction_max_wind_speed','direction_peak_wind_speed','max_wind_speed','days_with_fog']:
#     train_df[col]=imputer.fit_transform(train_df[[col]])
```

```python
#(2) using SimpleImputer
imputer = SimpleImputer(strategy='median')
train_df['year_built']= imputer.fit_transform(train_df[['year_built']])
train_df['energy_star_rating']= imputer.fit_transform(train_df[['energy_star_rating']])
train_df['direction_max_wind_speed']= imputer.fit_transform(train_df[['direction_max_wind_speed']])
train_df['direction_peak_wind_speed']= imputer.fit_transform(train_df[['direction_peak_wind_speed']])
train_df['max_wind_speed']= imputer.fit_transform(train_df[['max_wind_speed']])
train_df['days_with_fog']= imputer.fit_transform(train_df[['days_with_fog']])
```

```python
train_df.drop(['id'],axis=1,inplace=True)
```

```python
round(train_df.isna().sum()* 100 / len(train_df),2)
```

# Cleaning Data (Mistake Handling)

```python
#  year_built have mistake value "0"
train_df.loc[train_df['year_built']==0,'year_built']=train_df['year_built'].median()

#  'ELEVATION' have mistake minuse value "-6.4000e+00"

# change   minus values using abs(x)
train_df['ELEVATION']=train_df['ELEVATION'].apply(lambda x:abs(x))
```

# Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
steps = [('imputer', SimpleImputer(strategy='median')),
         ('scalar', StandardScaler()),
         ('selector',SelectPercentile(percentile=50, score_func=mutual_info_regression)),
         ('split',train_test_split(X, Y, test_size=0.2, random_state=2))
         ('model', XGBRegressor())]
```

```python
PipelineModel = Pipeline(steps)
PipelineModel.fit(X_test, y_test)

```
